```python
c, r = (2, 5)
x_origin = -1050
x_step = 1100
x_loc = [x_origin + (a * x_step) for a in range(c)]
xs = [x_loc[a % c] for a in range(r * c)]
assert len(xs) == c*r, 'Number of xs mismatch'
xs[:c]
```

```python
y_origin = 875
y_step = -550
y_loc = [y_origin + (a * y_step) for a in range(r)]
ys = [y_loc[a] for a in range(r) for _ in range(c)]
assert len(ys) == c*r, 'Number of ys mismatch'
ys[:r*c:c]
```

```python
coords = list(zip(xs, ys))
coords = [list(coord) for coord in coords]
assert len(coords) == c*r, 'Number of labels mismatch'
```

```python
column_label = ['', '']
row_label = ['1', '2', '3', '4', '5']
assert len(column_label) == c, 'Character list length mismatch'
assert len(row_label) == r, 'Numeral list length mismatch'

labels = []
for rl in row_label:
    for cl in column_label:
        labels.append(f'{cl}{rl}')
assert len(labels) == r*c, 'Number of label strings mismatch'
labels;
```

```python
map = []
for l, c in zip(labels, coords):
    map.append(
        {
            "x-size": 1000,
            "y-size": 450,
            "label": l,
            "origin": c
        }
    )
map
```

```python
# One grating type
x_size = 250
y_size = 205

x_origin = -1950 - 125
y_origin = 1940
# x_origin = -1815
# y_origin = 1940

x_step = 520
y_step = -215

alt_offset_x = -130

c, r = (8, 20)
```

```python
x_loc = [x_origin + (a * x_step) for a in range(c)]
x_loc_odd = [(x + alt_offset_x) for x in x_loc]
x_loc_odd.extend([x_loc_odd[-1] + x_step])
x_loc.extend(x_loc_odd)
xs = x_loc * 10
x_loc
```

```python
y_loc = [y_origin + (a * y_step) for a in range(r)]
ys = []
for row in range(r):
    if row % 2 == 0:
        count = c
    else:
        # count = c
        count = c + 1
    print(row, count, y_loc[row])
    for _ in range(count):
        ys.append(y_loc[row])
```

```python
coords = list(zip(xs, ys))
coords = [list(coord) for coord in coords]
print(len(coords))
```

```python
labels = []
for row in range(r):
    col = 0
    if row % 2 == 0:
        count = c
    else:
        count = c + 1
        # count = c
    for _ in range(count):
        grating_id = (row * 10) + col
        labels.append(f'{grating_id:03d}A')
        col += 1
labels
```

```python
map = []
for l, c in zip(labels, coords):
    map.append(
        {
            "x-size": 250,
            "y-size": 205,
            "label": l,
            "origin": c
        }
    )
map
```
